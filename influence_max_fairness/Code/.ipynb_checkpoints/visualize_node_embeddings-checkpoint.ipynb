```python
embedding_file_path = '../Data/Weibo/Embeddings/infector_source3.txt'
threshold = 0.999
```

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn import metrics
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from matplotlib.colors import LogNorm

from IPython.display import display
import warnings
warnings.filterwarnings("ignore") # To ignore warning messages while executing the code.

RANDOM_STATE = 0 # for reproducibility
```

```python
def embedding_matrix(path):
    embed_dim = [24748, 50] # total number of influencers in train_set.txt
    nodes = []
    f = open(path,"r")
    emb = np.zeros((embed_dim[0],embed_dim[1]), dtype=np.float)
    i=0
    for l in f:
        if "[" in l:
            combined = ""
        if "]" in l:
            combined = combined+" "+l.replace("\n","").replace("[","").replace("]","")
            parts = combined.split(":")
            nodes.append(int(parts[0]))
            emb[i] = np.asarray([float(p.strip()) for p in parts[1].split(" ") if p!=""],dtype=np.float)
            i+=1
        combined = combined+" "+l.replace("\n","").replace("[","").replace("]","")
    return nodes, emb
```

```python
def get_influencer_fairness(init_idx):
    influencer_fairness_read = open('../Data/Weibo/Init_Data/influencer_fairness.txt','r')
    gender_dict_read = {}
    for line in influencer_fairness_read:
        data = line.replace("\n","").split(",")
        gender_dict_read[data[0]] = data[1]
        
    influencer_fairness_list = []
    for n in init_idx:
        if n in init_idx:
            influencer_fairness_list.append(float(gender_dict_read[str(n)]))
        else:
            raise Exception("Sorry, {} is missing from train_cascade".format(n))
         
    return np.array(influencer_fairness_list)   
```

```python
%%time
nodes, emb = embedding_matrix(embedding_file_path)
influencer_fairness_array = get_influencer_fairness(nodes)
```

```python
train_data = emb.copy()
```

```python
def check_variance_explained(train_data):
    K, TotalVariance = [],[] # Creating empty lists for 'k' and corresponding 'Total Fraction of Explained Variance'
    for k in range(2, 51, 5): # Looping over different 'k' values ranging from 1 to 50
        pca = PCA(n_components = k, random_state = RANDOM_STATE) # Assigning PCA class to a variable
        pca.fit(train_data) # Fitting the train_data to PCA         
        K.append(k), TotalVariance.append(pca.explained_variance_ratio_.sum()) # Appeninding 'k' and corresponding 'total fraction of vairance explained' to the parent lists.
    
    # DataFrame representation
    df = pd.DataFrame({'K': np.array(K), 'Total Variance Explained': np.array(TotalVariance)}) # Creating a DatFrame with 'k' value and corresponding total fraction of vairance explained
    df.index += 1 # Dataframe index to start from 1 instead of 0
    display(df) # Display the data frame 
    
    #Plotting
    plt.plot(np.array(K), np.array(TotalVariance)) # Plotting 'k' vs. 'Total Fraction of Explained Variance'  
    plt.xlabel('Principal Component (k)', fontsize = 10) # xlabel of the plot 
    plt.ylabel('Total Fraction of Explained Variance', fontsize = 10) # ylabel of the plot
    plt.title('Principal Components(k) vs. Total Fraction of Explained Variance', fontsize = 15) # title of the plot
    plt.show() # plotting the figure    
    return None
```

```python
check_variance_explained(emb)
```

```python
def visualize_embeddings(train_data, train_labels, threshold):
    pca = PCA(n_components = 2, random_state = RANDOM_STATE) # Assigning PCA class (with n_components = 2)to a variable
    train_data_2dim = pca.fit_transform(train_data) # Fit_transforming the train_data on PCA  
    
    plt.figure(figsize=(20,10))
    plt.style.use('ggplot')
    plt.scatter(train_data_2dim[train_labels <= threshold, 0], train_data_2dim[train_labels <= threshold, 1], color = 'red', alpha = 0.1, label = 'Biased Influencers')
    plt.scatter(train_data_2dim[train_labels > threshold, 0], train_data_2dim[train_labels > threshold, 1], color = 'green', alpha = 0.9, label = 'Fair Influencers')
    plt.xlabel('Principal Component 1', fontsize = 18) # xlabel of the plot 
    plt.ylabel('Principal Component 2', fontsize = 18) # ylabel of the plot
    plt.legend(loc='best', shadow=True, scatterpoints=1, prop={'size': 14}) # legend location
    plt.title('Embeddings of train_set influencers (in 2 Principal Components)', fontsize = 20) # title of the plot
    plt.show()  # plotting the figure    
```

```python
visualize_embeddings(train_data, influencer_fairness_array, threshold)
```
